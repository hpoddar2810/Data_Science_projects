```
#genres = []
from ast import literal_eval
def getname(row, l):
  li = []
  row = literal_eval(row)
  for dic in row:
    name = dic['name']
    if name not in l:
      l.append(name)
    
    li.append(name)
  return li
```

```
def categorical_des(dataa):
  columns = dataa.columns
  index = ['Number of Nan Value', 'Number of categories', 'total row', 'Zeroes Count']#, 'Negative count']
  res = pd.DataFrame(columns=columns, index=index)

  for column in columns:
    res.loc[index[0], column] = len(dataa[dataa[column].isna()])
    res.loc[index[1], column] = dataa[column].nunique()
    res.loc[index[2], column] = len(dataa[column])
    res.loc[index[3], column] = len(dataa[(dataa[column].astype('str') == '0') | (dataa[column] == 0)])
    #res.loc[index[4], column] = len(data[data[column] < 0])
  
  return res
```
